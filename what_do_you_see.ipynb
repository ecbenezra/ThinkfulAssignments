```python
test_prep = df[df['test preparation course'] == 'completed']
test_prep
```

```python
no_prep = df[df['test preparation course'] == 'none']
no_prep
```

```python
#Normalized histograms that show the math score distribution between the students 
#who took a 'test preparation course' (blue) and those who did not (yellow). It could be inferred from this data that
#students who took the test preparation course performed better on the exam than their counterparts without the course

plt.hist(test_prep['math score'], density=True, ls='dashed', alpha = 0.5, lw=3, color= 'blue')
plt.hist(no_prep['math score'], density=True, ls='dotted', alpha = 0.5, lw=3, color= 'yellow')
plt.show()
```

```python
#Side by side box plots show that students who took the test prep course had fewer outlier test scores, with only a few
#fliers below the lower whisker. Median score ~70 for test prep course students. There was a lower median score for
#students who did not take the test prep course, but not by much. There are more fliers below the lower whisker. This
#could mean that, without taking into account socioeconomic or other factors for testing success, students who took 
#the test prep course prior to the exam had slightly higher scores than their counterparts. There were no students
#who took the test prep course who scored below 20 points, whereas there were several who did not take the test prep
#course who scored below 20 points.

plt.figure(figsize=(15, 10))

plt.subplot(1, 2, 1)
plt.boxplot(test_prep['math score'])
plt.title('Test Prep Course')

plt.subplot(1, 2, 2)
plt.boxplot(no_prep['math score'])
plt.title('No Test Prep Course')
plt.show()
```
