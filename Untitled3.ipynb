```python
for col in df.columns:
    df[col] = df[col].astype(str)
    

scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
```

```python
#Make a new column, text, that will display the state, the name of the product, and the value for the row

df['text'] = df['State'] 

#+ '<br>' +\
#    'Males '+df['Males']+' Females '+df['Females']+'<br>'+\
#    'Rural '+df['Rural']+' Urban ' + df['Urban']

df
```

```python
#dict() creates a dictionary
#type='choropleth' indicates the type of map
#colorscale = scl indicates the colors we set earlier
#locations = df['code'] indicates the locations are state codes
#z = df['total exports'].astype(float) turns the total export numbers back into a float from a string
#locationmode = 'USA-states', draws from an inner map thing, I guess
#text = df['text'] will display what we set as the new 'text' column above
#marker = dict(
#line = dict(
#color = 'rgb(255,255,255)'
#width = 2
#)),
#colorbar = dict(
# title = "Millions USD")
#)]

#This is the data that fills out the map

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df['State'],
        locationmode='ISO-3',
        z = df['Persons'].astype(float),
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Population")
        ) ]



layout = dict(
        title = 'Population of India',
        geo = dict(
            scope='asia',
            ))
```

```python
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```

```python
df = pd.read_csv('all.csv')

for col in df.columns:
    df[col] = df[col].astype(str)


scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]

df['text'] = df['State'] 

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df['State'],
        locationmode='ISO-3',
        z = df['Males'],
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Population")
        ) ]



layout = dict(
        title = 'Population of India',
        geo = dict(
            scope='asia',
            
            ))
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```
